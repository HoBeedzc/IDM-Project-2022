```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
```

```python
# Load and prepare the data.
data = pd.read_excel("./Data/AR-CatalogCrossSell.xls",sheet_name="Data")
data
```

```python
# Drop null values and filter out all zero values
del data["Unnamed: 1"]
del data["Customer Number"]
del data["Health Products Division"] # all in this column is 1. No meaningful value.
data = data.dropna()
data
```

```python
# Use an association rules analysis
#利用 Apriori 找出频繁项集
freq = apriori(data, min_support=0.1, use_colnames=True)
freq.sort_values(by="support")
```
